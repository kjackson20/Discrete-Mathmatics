**Conjunction: r ⋀ p**

The conjunction of r and p is true when both r and p are true. Since statements r and p are true, r ⋀ p is true, otherwise false. 

Below is a truth table for "and", also known as a conjunction.

```
  r   |   p   |   r ⋀ p
________________________
 True    True     True
 True    False    False
 False   True     False
 False   False    False
```

```
  r   |   ¬q   |  r ⋁ ¬q
_________________________

 False   True      True
```

The implication of r and p is short hand for the statement "if r then p". Implication requires p to be true whenever r is true. If r is not true, then p can be any value. Implication is false when r is true and p is false. The implication of r and p is short hadn for the statement "if r then p".

Below is a truth table for "if, then", also known as implication.



```
  r   |   p   |   r ⇒ p
________________________
 True    True     True
 True    False    False
 False   True     True
 False   False    True
```



````
def implication(r, p):
 return not (r) or p
 print("r    p    a")
for r in [True, False]:
    for p in [True, False]:
      a = implication(r, p)
      print(r, p, a)
````
