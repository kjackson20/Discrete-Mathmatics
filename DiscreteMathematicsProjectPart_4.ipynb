When order does not matter, the number of possible outcomes will be the number of combinations. 
The formula that we could use for this is:

$$
\binom{n+k-1}{k-1} = \binom{n+k-1}{n}
$$

If we were not allowed to select an item multiple times, the the permutation would be the number of possible outcomes where order would matter.

When order matters, then the 10 events can be arranged in 10 ways.

We have to select 10 simulations of events out of 20 possible outcomes.

```
from itertools import combinations_with_replacement

print(list(combinations_with_replacement(possibleOutcomes,10)))
```

```
len(list(combinations_with_replacement(possibleOutcomes,10)))
```
